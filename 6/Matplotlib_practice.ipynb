## üìå Grid, Axes, and Labels

[legend Ï∞∏Í≥†ÏûêÎ£å](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.legend.html#matplotlib.axes.Axes.legend)

```python
x = np.arange(1,8) # 1Î∂ÄÌÑ∞ 7ÍπåÏßÄ
y1 = -x**2
y2 = -x**3
y3 = -2*x
y4 = -2**x
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.grid() # Í≤©Ïûê 
# axes Ï°∞Ï†ï
plt.xlim([4,6]) # xÏ∂ï ÏûêÎ•¥Í∏∞
plt.ylim([-150,0]) # yÏ∂ï ÏûêÎ•¥Í∏∞ 
plt.show() 
```

```python
x = np.arange(1,8) # 1Î∂ÄÌÑ∞ 7ÍπåÏßÄ
y1 = -x**2
y2 = -x**3
y3 = -2*x
y4 = -2**x
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.grid() # Í≤©Ïûê 
# axes Ï°∞Ï†ï
plt.xlim([4,6]) # xÏ∂ï ÏûêÎ•¥Í∏∞
plt.ylim([-150,0]) # yÏ∂ï ÏûêÎ•¥Í∏∞ 
# labels
plt.ylabel('y = f(x)')
plt.xlabel('x = np.arange(1,8)')
plt.legend() # warning Î∞úÏÉù. 
plt.show() 
```

```python
x = np.arange(1,8) # 1Î∂ÄÌÑ∞ 7ÍπåÏßÄ
y1 = -x**2
y2 = -x**3
y3 = -2*x
y4 = -2**x
plt.plot(x,y1, label = 'y1 = -x**2')
plt.plot(x,y2, label = 'y2 = -x**3')
plt.plot(x,y3, label = 'y3 = -2*x')
plt.plot(x,y4, label = 'y4 = -2**x')
plt.grid() # Í≤©Ïûê 
# axes Ï°∞Ï†ï
plt.xlim([4,6]) # xÏ∂ï ÏûêÎ•¥Í∏∞
plt.ylim([-150,0]) # yÏ∂ï ÏûêÎ•¥Í∏∞ 
# labels
plt.ylabel('y = f(x)')
plt.xlabel('x = np.arange(1,8)')
plt.title('Simple Plot Demo')
plt.legend() # warning Ìï¥Í≤∞.
plt.show() 
```

```python
x = np.arange(1,8) # 1Î∂ÄÌÑ∞ 7ÍπåÏßÄ
y1 = -x**2
y2 = -x**3
y3 = -2*x
y4 = -2**x
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.grid() # Í≤©Ïûê 
# axes Ï°∞Ï†ï
plt.xlim([4,6]) # xÏ∂ï ÏûêÎ•¥Í∏∞
plt.ylim([-150,0]) # yÏ∂ï ÏûêÎ•¥Í∏∞ 
# labels
plt.ylabel('y = f(x)')
plt.xlabel('x = np.arange(1,8)')
plt.title('Simple Plot Demo')
plt.legend([
    'y1 = -x**2',
    'y2 = -x**3',
    'y3 = -2*x',
    'y4 = -2**x'], loc = 'lower center') # Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Îäî ÏàúÏÑú Í∞ôÏïÑÏïºÌï®. 
plt.show() 
```

## üìå Colors, Styles and Markers

```python
# ÎßâÎåÄÍ∑∏ÎûòÌîÑ
x = [1,2,3,4,5,6]
y = [4,5,3,1,6,7]
plt.bar(x,y,color="green")
plt.show()
```

```python
x = [1,2,3,4]
y = [4,5,3,1]
bar_labels = ['red','blue','red','orange']
bar_colors = ['tab:red','tab:blue','tab:red', 'tab:orange']
plt.bar(x,y,label = bar_labels, color=bar_colors)
plt.title('hello')
plt.ylabel('ylabel')
plt.legend()
plt.show()
# Í≤∞Í≥ºÎ•º Î≥¥Î©¥, ÎùºÎ≤®Ïù¥ Ï†úÎåÄÎ°ú ÏïàÎÇòÏò§ÎäîÎç∞ Î≤ÑÏ†ÑÏù¥ 3.2.2 Î≤ÑÏ†ÑÏù¥ÎùºÏÑú Í∑∏Î†áÎã§.
```

```python
x = [1,2,3,4]
y = [4,5,3,1]
bar_labels = ['red','blue','red','orange']
bar_colors = ['tab:red','tab:blue','tab:red', 'tab:orange']

plt.bar(x[0],y[0],label = bar_labels[0], color=bar_colors[0])
plt.bar(x[1],y[1],label = bar_labels[1], color=bar_colors[1])
plt.bar(x[2],y[2],label = bar_labels[2], color=bar_colors[2])
plt.bar(x[3],y[3],label = bar_labels[3], color=bar_colors[3])

plt.title('hello')
plt.ylabel('ylabel')
plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y = [4,5,3,1]
bar_labels = ['red','blue','red','orange']
bar_colors = ['tab:red','tab:blue','tab:red', 'tab:orange']

# Î∞òÎ≥µÎ¨∏ÏúºÎ°ú ÎèåÎ¶¨Í∏∞
for i,_ in enumerate(x):
  plt.bar(x[i],y[i],label = bar_labels[i], color=bar_colors[i])

plt.title('hello')
plt.ylabel('ylabel')
plt.legend()
plt.show()
```
